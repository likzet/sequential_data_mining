### List

````python
b = [1, 3, 7, 2, 2]
print(b)
# select a part of the list (indexing from zero)
print(b[2])
print(b[:3])
print(b[-1])
print(b[-3:-1])
print(b[1:5:2])
# for multidimensional array: not very useful
````

````python
# we can change a list
b[-1] = 10000
print(b)
print(b[2])
````

````python
# but we can't change the total size in this way
b[10] = -1
print(b)
````

````python
# we can modify the list using a specific function
print(b)
b.insert(1, -1)
print(b)
````

````python
# ...also we can add an element to the end
b.append(-1)
print(b)
````

````python
# concatenate two list (types can be arbitrary)
c1 = [1, 3, 5]
c2 = ["bank", "device", 100]
print(c1 + c2)
````

````python
# можно вызывать встроенные функции
print(len(b))
print(b)
print(sorted(b))
print(sorted(c1 + c2))
````

````python
# call other methods...
print(b.count(-1))
````

````python
# We can get all values from an iterator (e.g. list)
print([i for i in range(4)])
print([i ** 2 for i in range(4)])
````

## NumPy - library for processing of multidimensional arrays

````python
import numpy as np
````

### Comparison to lists

````python
# Let us create an array and apply a couple of operations to it
a = np.array([[0, 1, 2], [3, 4, 6]])
print(a.shape)
print('a: \n', a)
print('a + 1: \n', a + 1)
print('a[0, 1]: \n', a[0, 1])
````

````python
n = 1000
list_array1 = [i for i in range(n)]
list_array2 = [i / 2. for i in range(n)]
res_list_array = [0 for i in range(n)]
````

````python
# to measure time of execution of a specific operation we use magic command timeit
%%timeit
for i in range(n):
    res_list_array = list_array1[i] + list_array2[i]
````

````python
np_array1 = np.array(list_array1)
np_array2 = np.array(list_array2)
%timeit np_array1 + np_array2
````

````python
# можно создавать массивы нулей или единиц заданного размера 
a = np.zeros((3, 2))
print(a)
b = np.ones((3, 2))
print(b)
````

````python
# или создавать масссивы из случайных элементов
A = np.random.rand(4, 3)
print(A)
print(A.T)
````

````python
print(np.max(A), A.max())
print(np.min(A), A.min())
print(np.sum(A), A.sum())
````

````python
# indexing are even richer for np.arrays
print(A[:3, 0:1])
print(A[0, 1])
print(A[1:3, 2:])
````

````python
# also linear algebra are easily applicable
print(np.linalg.norm(A, 1))
print(np.linalg.norm(A, 2))
print(np.linalg.norm(A, "fro"))
````

````python
# multiply matrix by a vector
x = np.random.randn(A.shape[1], 1)
print(A @ x)
````

````python
# Multiply rows of a matrix by a number
x = np.random.randint(0, 10, size=(A.shape[0], 1))
print(A) 
print(x)
print(A * x)
````

````python
# Solving a system of linear equations
A = np.random.randn(3, 3)
b = np.random.randn(3)
x = np.linalg.solve(A, b)
print(np.linalg.norm(A @ x - b))
````

### Exercise

1. Use singular decomposition for the image available under ```https://drive.google.com/uc?id=1as5__catiMerhA0OASc7MX89olzttSq2``` below using the function ```np.linalg.svd```
2. Look at the singular values
3. Use only 10, 50, 100, or 150 singular values and vectors to reconstruct the image

````python
import imageio
import matplotlib.pyplot as plt
import numpy as np
````

````python
skoltech_logo = imageio.imread("https://drive.google.com/uc?id=1as5__catiMerhA0OASc7MX89olzttSq2")
greyscale_coefficients = [0.2989, 0.5870, 0.1140]
skoltech_logo_greyscale = (greyscale_coefficients[0] * skoltech_logo[:, :, 0] + 
                           greyscale_coefficients[1] * skoltech_logo[:, :, 1] + 
                           greyscale_coefficients[2] * skoltech_logo[:, :, 2])
plt.imshow(skoltech_logo_greyscale[:, :])
_ = plt.axis("off")
````

````python
# YOUR CODE HERE
````

### Exercise 1 Solution

````python
U, D, V = np.linalg.svd(skoltech_logo_greyscale)
````

### Exercise 2 Solution

````python
print(np.shape(U), np.shape(D), np.shape(V))
````

### Exercise 3 Solution

````python
cropped_D = np.array(D)
cropped_D[10:] = 0
decomposed_image = (U * cropped_D) @ V[:len(cropped_D), :]
````

````python
plt.imshow(decomposed_image)
_ = plt.axis("off")
````

## Matplotlib - the most popular plotting library in Python

````python
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
titanic = np.genfromtxt('https://drive.google.com/uc?id=1Tb63anYEk8P1KktDfe7Shgq2AfYuSCx1', skip_header=1, 
                       delimiter=',')
print(titanic.shape)
````

````python
age = titanic[:, 6]
print(age[:100])
````

````python
print(age[~np.isnan(age)].min())
print(age[~np.isnan(age)].max())
print(age[~np.isnan(age)].mean())
````

````python
plt.hist(age[~np.isnan(age)], bins=100);
````

````python
fare = titanic[:, -3]
fare[:100]
````

````python
# A plot
plt.figure(figsize=(10, 8))
plt.plot(fare[~np.isnan(fare)], '-o')

# Beautification
plt.xlabel('Index', fontsize=24)
plt.ylabel('Fare', fontsize=24)

plt.rc('xtick', labelsize=18)
plt.rc('ytick', labelsize=18)

# plt.axis([0, 90, 20, 95])
plt.grid(True)

# Good way to save the figure (use .pdf instead to get a vector image)
# plt.gcf().set_facecolor('white')
# plt.savefig('well_reserve.png', dpi=400, bbox_inches='tight')
````

````python
plt.bar([1, 2, 3], [np.sum(pclass == 1), np.sum(pclass == 2), np.sum(pclass == 3) ])
_ = plt.xticks([1, 2, 3], ["First", "Second", "Third"])
````
